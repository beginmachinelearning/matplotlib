```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
data_for_first_plot = np.arange(20)
```

```python
data_for_first_plot
```

```python
plt.plot(data_for_first_plot)
```

## Let us customize our graph

```python
plt.plot(data_for_first_plot, color='g')
#What happened? Where is my graph? Look Above
```

## Let us create a new graph

```python
plt.figure('red_o_dashed') #Named it red_o_dashed
plt.plot(data_for_first_plot, color='r', linestyle='--', marker='o')
```

## Doing the same together

```python
plt.figure('red_o_dotted')
plt.plot(data_for_first_plot, 'ro:')
```

## Go back and run the red_o_dashed code again 

```python
plt.plot(data_for_first_plot, 'yo--')
```
