```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

# Let us have 20 people of different ages


```python
age_set=[20, 3, 5, 33, 41, 92, 74, 54, 80, 11, 26, 37, 64, 59, 33, 73, 26, 38, 44, 7]
```

# Default partitions(bins) is 10, each shows number of people in that age bracket


```python
plt.figure()
plt.hist(age_set)
```

# This time we break it down to 20 partitions

```python
plt.figure()
plt.hist(age_set, bins=20)
```

# Add Colors and opacity

```python
plt.figure()
plt.hist(age_set, color='g', alpha=0.5)
```
