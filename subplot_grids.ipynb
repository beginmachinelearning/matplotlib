## How does the subplot numbering work. .add_subplot(xyz) where x is number of rows, y is number of cols, z is the sequential order
### This time let us create a 2 x 2 grid

```python

figure_2x2=plt.figure(figsize=(10, 4))



## We create 4 slots, by breaking the whole figure into 2 x 2 grid. We use 4 slots, leaving the 3rd unused
data_for_first_plot_subplot=figure_2x2.add_subplot(221)
age_set_subplot=figure_2x2.add_subplot(222)
houseprice_year_subplot=figure_2x2.add_subplot(224)
```

## Where is the data for the graph? I see only boxes
### Add subplots, only creates slots for the plot. You have to use .plot, .hist, .scatter to actually plot the graph

```python
figure_2x2_split=plt.figure(figsize=(10, 4))


## We create 4 slots, by breaking the whole figure into 2 x 2 grid. Each addsubplot command breaks the graph on its own
data_for_first_plot_subplot=figure_2x2_split.add_subplot(211) #This breaks it into 2 rows, 1 column, occupying the first slot
age_set_subplot=figure_2x2_split.add_subplot(223) #This breaks it into 2 rows, 2 column, occupying the third slot
houseprice_year_subplot=figure_2x2_split.add_subplot(224)  #This breaks it into 2 rows, 2 column, occupying the 4th slot
```

```python
figure_irregular=plt.figure(figsize=(10, 4))


## We create irregular slot for each subplot
data_for_first_plot_subplot=figure_irregular.add_subplot(211) #This breaks it into 2 rows, 1 column, occupying the first slot as before
age_set_subplot=figure_irregular.add_subplot(223) #This breaks it into 2 rows, 2 column, occupying the third slot
houseprice_year_subplot=figure_irregular.add_subplot(257)  #This breaks it into 2 rows, 5 columns, making it a 10 slotted 2x5 grid, occupying the 7th slot which is row 2, column 2 leading to overlap
```
