```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
houseprice_year=np.array([[2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010], 
                          [169191,182863,198087,224567,249191,276221,305637,322634,305269,242033,228268]], np.int32)
                          
houseprice_year=np.transpose(houseprice_year)  
figure=plt.figure(figsize=(10, 5))
houseprice_year_subplot=figure.add_subplot(111)
houseprice_year_subplot.scatter(houseprice_year[:, 0], houseprice_year[:, 1])
```

```python
houseprice_year_subplot.set_title("Scatter plot of Price vs Year")
```

```python
houseprice_year_subplot.set_xticks([2000, 2004, 2008, 2012])
```

```python
houseprice_year_subplot.set_xticklabels(["Year 2000", "4 Years Later", "8 Years Later", "12 Years Later"])
```

```python
houseprice_year_subplot.set_xticklabels(["Year 2000", "4 Years Later", "8 Years Later", "12 Years Later"], rotation=30, fontsize='small')
```

```python
houseprice_year_subplot.set_xticklabels(["Year 2000", "4 Years Later", "8 Years Later", "12 Years Later"], rotation=0, fontsize='small')
```

```python
houseprice_year_subplot.set_xlabel('Years')
houseprice_year_subplot.set_ylabel('Cost in USD')
```

```python
houseprice_year_subplot.plot(np.random.rand(10000).cumsum(), 'k', label='one')
```

```python
figure_legend=plt.figure(figsize=(10, 5))
houseprice_year_subplot=figure_legend.add_subplot(111)
houseprice_year_subplot.scatter(houseprice_year[:, 0], houseprice_year[:, 1])
houseprice_year_subplot.legend(["Price"])
### Note that the legend needs an array of variabled to be passed
```

```python
random_ints1=random.sample(range(50000, 250000), 11)

random_ints2=random.sample(range(50000, 250000), 11)
```

```python
houseprice_year_subplot.scatter(houseprice_year[:, 0], random_ints1 )


houseprice_year_subplot.scatter(houseprice_year[:, 0], random_ints2)
houseprice_year_subplot.legend(["Price", "New Scatter", "Yet Another Scatter"])
```
